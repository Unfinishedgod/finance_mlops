```python
from pykrx import stock
from pykrx import bond
import pandas as pd

type_list = ["금융투자","보험","투신","사모","은행","기타금융","연기금","기관합계","기타법인","개인","외국인","기타외국인","전체"]

total_df = pd.DataFrame()

for type_nm in type_list:
    df = stock.get_market_net_purchases_of_equities("20230102", "20230102", "ALL", type_nm)
    df['type'] = type_nm
    total_df = pd.concat([total_df, df])

    total_df = total_df.reset_index()
```
